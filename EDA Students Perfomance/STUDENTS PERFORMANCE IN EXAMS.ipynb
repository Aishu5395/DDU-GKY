# Data visualisation

# Univariate Analysis

## Histogram and  KDE plot

### Data showing writing score of students

```python
sns.histplot(data=data.writing_score,color='g',kde=True)
plt.show()
```

##### The average score of writing score is between 60 and 80.

## Box plot

### Data showing reading score of students

```python
sns.boxplot(data=data,y='reading_score')
plt.show()
```

##### The avg of reading score is from 60 to 80  and  there is  an outlier that is less than 20 

## Bar plot

### Data showing parental level of education

```python
data.parents_edu_level.value_counts()
```

```python
data.parents_edu_level.value_counts().plot.bar(figsize=(8,6),cmap='Paired');
```

##### Parents education from some college is highest associate's degree is also same.whereas master's degree is lowest

## Count plot

### Data showing number of students each ethnicity

```python
data.ethnicity.value_counts()
```

```python
plt.figure(figsize=(6,4))
plt.title('Number of students each ethnicity')
sns.countplot(data=data, x='ethnicity', palette='RdPu')
plt.show();
```

##### Students that belong to group C are highest, whereas group A is the lowest.

## Pie plot

### Data showing the type of lunch students prefer the most

```python
data.lunch.value_counts()
```

```python
data.lunch.value_counts().plot.pie(autopct ='%1.2f%%',colors=['#FFC0CB','#ADD8E6'], \
              explode=[0.1,0],legend = True);
```

##### Students who got standard lunch are more than students who got free/reduced lunch.

### Data showing gender distibution

```python
data.gender.value_counts()
```

```python
data.gender.value_counts().plot.pie(autopct ='%1.2f%%',explode=[0.1,0],\
             colors=['#FFA07A','#CCCCFF'],legend = True);
```

##### Female students are more in number than male students.

# Multi variate analysis

## Catplot

```python
sns.catplot(data=data,x='ethnicity',y='math_score',hue='gender')
plt.show()
```

##### Group E got more scores in maths exam

## Box plot

### Data showing lunch and mathscore of students

```python
sns.boxplot(data = data, x = 'lunch', y= 'math_score',palette ='BuPu')
plt.show()
```

##### Students who consume std lunch more scored compared to free/reduced lunch

## Scatter plot

### Data showing gender wise score distribution of maths and reading

```python
sns.scatterplot(data=data,x='math_score',y='reading_score',hue='gender')
plt.show()
```

##### The average score is between 60 to 80.Female scored more marks compared to male students

### Rel plot

#### Data showing relationship between reading score and writing score

```python
sns.relplot(data, x="reading_score", y="writing_score", height=3.5, aspect=1.2, size=1, color="red",legend=False)
plt.show()
```
